```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#prepare sample training data
X=2*np.random.rand(1000,1)
y=4+3*X+np.random.rand(1000,1)
plt.plot(X,y,"b.")
```

```python
X_b=np.c_[np.ones((1000,1)),X] # 1 as bias is added
```

Do Gradient Descent

```python
learning_rate=0.1
num_training_samples=X.shape[0]
iterations=1000
```

```python
theta=np.random.randn(2,1)
print('Initial Theta ')
print(theta)
```

```python
for iter in range(iterations):
    # get the gradient of the cost function(partial derivative) which gives the slope which points uphill
    gradient=2/num_training_samples*X_b.T.dot(X_b.dot(theta)-y) 
    # follow downhill so subtract the gradient
    theta=theta-learning_rate*gradient
print(theta)    
```

```python
# Predict on new samples using the estimated theta
X_new=np.array([[0],[2]])
X_new_b=np.c_[np.ones((2,1)),X_new]
y_predict=X_new_b.dot(theta)
plt.plot(X,y,"b.")
plt.plot(X_new,y_predict,'r-')
```
