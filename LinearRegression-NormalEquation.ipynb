```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Prepare a simple linear Dataset

```python
X=2*np.random.rand(1000,1)
y=4+3*X+np.random.rand(1000,1)
plt.plot(X,y,"b.")
```

```python
X_b=np.c_[np.ones((1000,1)),X] # 1 as bias is added
```

```python
theta_best=np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T.dot(y)) # use normal equation to find params that minimizes the equation
```

```python
theta_best
```

Predict using estimated theta

```python
X_new=np.array([[0],[2]])
X_new_b=np.c_[np.ones((2,1)),X_new]
```

```python
y_predict=X_new_b.dot(theta_best)
```

```python
plt.plot(X,y,"b.")
plt.plot(X_new,y_predict,'r-')
```
